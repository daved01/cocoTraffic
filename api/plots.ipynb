# Plots for the 4 datasets

```python
import numpy as np
import matplotlib.pyplot as plt
import json
%matplotlib inline
plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
def plot_dataset(path, filename, plot_name):
    # Load file
    f = open(path+filename)
    anns = json.load(f)

    # Extract instances per category
    stats = dict()

    for ann in anns['annotations']:
        if ann['category_id'] in stats:
            stats[ann['category_id']] += 1
        else:
            stats[ann['category_id']] = 1

    print("Number of images: {}".format(len(anns["images"])))
    print("Number of annotations: {}".format(len(anns["annotations"])))

    # Get category names
    cats = {x['id']:x['name'] for x in anns['categories']}
    labels = list(sorted(stats.keys()))
    names = [cats[x] for x in labels]

    plt.figure()
    plt.bar(names, stats.values(), color='blue')
    plt.xticks(rotation=90)
    path_fig = "../plots/"
    plt.savefig(plot_name + ".png")
    plt.show()
```

## 1. Dataset - COCO Traffic Lights

```python
path = '../annotations/01_coco_traffic_lights/'
filename = 'instances_traffic_lights.json'

plot_dataset(path, filename, "1_coco_traffic_lights")
```

## 2. COCO Traffic Full

```python
path = '../annotations/01_coco_traffic_full/'
filename = 'instances_train2017traffic.json'
plot_dataset(path, filename, "1_coco_traffic_full_train")
```

```python
path = '../annotations/01_coco_traffic_full/'
filename = 'instances_val2017traffic.json'
plot_dataset(path, filename, "1_coco_traffic_full_val")
```
